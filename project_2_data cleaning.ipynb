```python
# Import dependencies
import pandas as pd
```

```python
# Read in files
identity_df = pd.read_csv('datafiles/train_identity.csv')
transaction_df = pd.read_csv('datafiles/train_transaction.csv')
```

```python
# Check identity_df
identity_df
```

```python
# Check transaction_df
transaction_df
```

```python
# Merge identity_df and transaction_df on TransactionID
transaction_merged = pd.merge(identity_df, transaction_df, on=["TransactionID"])

transaction_merged

#transaction_merged = transaction_merged.drop_duplicates()
# All rows are unique
```

```python
# Clean data and remove colomns with more than 10% missing values
# Calculate the percentage of missing values for each column
missing_percentage = transaction_merged.isnull().mean() * 100

# Reset index to convert the Series to a DataFrame
df_missing_percentage = missing_percentage.reset_index()
df_missing_percentage.columns = ['Column', 'Missing Percentage']

# Identify columns with more than 10% missing values
columns_above_10 = df_missing_percentage[df_missing_percentage['Missing Percentage'] > 10]

# Display columns with more than 10% missing values
drop_columns = columns_above_10 ['Column'].tolist()

df_dropped = transaction_merged.drop(columns=drop_columns)

df_dropped
```

```python
# Identify non-numeric columns
non_numeric_cols = df_dropped.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)
```

```python
# Handle non-numeric data
df_dropped_dummies = pd.get_dummies(df_dropped, columns=non_numeric_cols)
df_dropped_dummies
```

```python
# Check again non-numeric columns
non_numeric_cols = df_dropped_dummies.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)
```
